<!-- ## O Danych

Zbiór danych zawiera informacje o trzech gatunkach irysów: Iris setosa, Iris versicolor, i Iris virginica.

Dane obejmują pomiary czterech cech: długość i szerokość działki kielicha oraz długość i szerokość płatka.

Każdy wiersz w zbiorze danych reprezentuje pojedynczy kwiat, a wartości pomiarów są podane w centymetrach.

Zbiór składa się z 150 próbek, po 50 dla każdego gatunku, i jest szeroko stosowany jako podstawowy zbiór danych do
testowania algorytmów klasyfikacji oraz w nauce o danych i uczeniu maszynowym.

Kolumny:
* **długość kielicha (sepal length)** - Długość kielicha w cm
* **szerokość kielicha (sepal width)** - Szerokość kielicha w cm
* **długość płatka (petal length)** - Długość płatka w cm
* **szerokość płatka (petal width)** - Szerokość płatka w cm
* **klasa (class)** - Klasa irysa (setosa, versicolor, virginica) -->

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('25__iris.csv', sep=",")
```

# Eksploracyjna analizę danych o Irysach


## 1.Badamy poszczególne kolumny

### 1.1 Sprawdźmy ile jest poszczególnych irysów osiągających daną wartości miary poszczególnych elementów ich budowy

```python
df.hist(bins=100);
```

Obserwujemy że nawięcej irysów osiąga:
- długość kielicha równą `5cm`, 
- szerokość kielicha równą `3 cm`,
- długość płatka równą około `1,1 cm`,
- szerokość płatka równą około `0,2 cm`

### 1.2 Pogrupujemy teraz irysy na wszytskie dostępne gatunki i sprawdźmy poszczególne średie,mediany,min i max wartości elementów ich budowy

```python
irys_mean_df=df.groupby("klasa (class)").agg({
    "długość kielicha (sepal length)":["mean"],
    "szerokość kielicha (sepal width)":["mean"],
    "długość płatka (petal length)":["mean"],
    "szerokość płatka (petal width)":["mean"]
})

# zminna obrazująca pogrupowane średnie elementów budowy Irysów
```

```python
irys_median_df=df.groupby("klasa (class)").agg({
    "długość kielicha (sepal length)":["median"],
    "szerokość kielicha (sepal width)":["median"],
    "długość płatka (petal length)":["median"],
    "szerokość płatka (petal width)":["median"]
})
#irys_median_df
# # zminna obrazująca pogrupowane mediany elementów budowy Irysów
```

<!-- #### 1.Dane tabearyczne poszczególnych gatunków -->

```python
irys_mean_median_df=df.groupby("klasa (class)").agg({
    "długość kielicha (sepal length)":["mean","median"],
    "szerokość kielicha (sepal width)":["mean","median"],
    "długość płatka (petal length)":["mean","median"],
    "szerokość płatka (petal width)":["mean","median"]
})
# irys_mean_median_df
# # zminna obrazująca pogrupowane średnie i mediany elementów budowy Irysów
```
